#### Необходимо

1. выполнить анализ и охарактеризовать клиентский портфель организации

##### Из графиков видно, что у клиентов с одобренным займом просрочка минимальна, но количество микрокредитов достаточно большое. Возможно, клиенты перекредитовываются, чтобы погасить предыдущий микрокредит с минимальной переплатой. Однако, для более полного представления о клиентах, нам не хватает бинарных данных, таких как наличие работы, собственности, гендер и прочих. Эти дополнительные данные помогли бы лучше понять профиль таких клиентов.


```python
#Посмотроим на коррелирущие данные
def get_redundant_pairs(df):
    pairs_to_drop = set()
    cols = all_reject_data.columns
    for i in range(0, all_reject_data.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = all_reject_data.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(all_reject_data)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Absolute Correlations")
print(get_top_abs_correlations(all_reject_data, 30))
```

##### Много признаков, которые имеют высокую корреляцию, это может вызвать проблему мультиколлинеарности в линейных моделях.

#### Приступим к обучению модели

```python
X = new_training_data.drop('target', axis=1)
y = new_training_data['target']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=0,
                                                    shuffle=True,
                                                    test_size=0.1)
```

##### Так как я не стал удалять коррелирующие столбцы не будем использовать линейные модели такие как Logistic Regression и SVM

```python
# Список моделей c дефолтными параметрами, ограничим только глубину для обучения
models = [
    ('Random Forest', RandomForestClassifier(random_state=1, max_depth = 3)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=1, max_depth = 3)),
    ('XGBoost', xgb.XGBClassifier(random_state=1, max_depth = 3))
]

# Обучение моделей и вычисление ROC AUC для контроля переобучения модели и валидации
for model_name, model in models:
    model.fit(X_train, y_train)
    y_pred_train = model.predict_proba(X_train)[:, 1]
    y_pred_test = model.predict_proba(X_test)[:, 1]
    
    roc_auc_train = roc_auc_score(y_train, y_pred_train)
    roc_auc_test = roc_auc_score(y_test, y_pred_test)
    cm = confusion_matrix(y_test, model.predict(X_test), labels=model.classes_)

    cmp = ConfusionMatrixDisplay(confusion_matrix=cm)
    fig, ax = plt.subplots(figsize=(7,4))
    cmp.plot(ax=ax)

    plt.show()
    
    print(f"{model_name} ROC AUC_train: {roc_auc_train}")
    print(f"{model_name} ROC AUC_test: {roc_auc_test}")
```

#### Наша цель - разработать модель, которая будет одобрять не менее 35% клиентов при банкротстве среди одобренных не выше 15%. Для достижения этой цели, нам нужно сделать модель более склонной к предсказанию банкротств.

#### Для этого мы применим технику, которая позволяет нам создать искусственное условие. Мы воспользуемся параметром class_weight и увеличим вес для target=1 (банкротства).

#### Далее приступим к настройке параметров и дотюниванию модели XGBoost, чтобы получить наилучшие прогнозы и соответствовать требованиям заказчика. 

#### Клиенты которым рекомендован заем будут в классе 0 и им будет соответствовать True Negative и False Negative (в сумме более 35% от всех заявок), False Negative - клиенты которым одобрен заем, но по итогу они стали банкротами (необходимо менее 15%).



```python
class_weight = {0: 1, 1: 3.5}

model = xgb.XGBClassifier(learning_rate=0.1, n_estimators=150 ,max_depth = 4, subsample = 0.4, scale_pos_weight=class_weight[1])

model.fit(X_train, y_train)


tn, fp, fn, tp = confusion_matrix(y_test, model.predict(X_test)).ravel()

print(f'True Negative errors: {tn}') 
print(f'False Positive errors: {fp}') 
print(f'False Negative errors: {fn}') 
print(f'True Positive errors: {tn + fn}') 

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

print(f"Train ROC-AUC score: {roc_auc_score(y_train, model.predict_proba(X_train)[:, 1])}")
print(f"Test ROC-AUC score: {roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])}")

cm = confusion_matrix(y_test, model.predict(X_test), labels=model.classes_)

cmp = ConfusionMatrixDisplay(confusion_matrix=cm)
fig, ax = plt.subplots(figsize=(7,4))
cmp.plot(ax=ax)

plt.show()
```

### По итогу перебору параметров GridSearchCV и максимизации precision для 0 класса, подбору нужного class_weight, обучена модель удовлетворяющая заказчика. 

```python
total_loans = tn + fp + fn + tp
print('Ключевые метрики:')
print(f'Количество выданных займов: {tn + fn}, процент выданных займов от общего числа заявок: {((tn + fn) / total_loans) * 100:.1f}%')
print(f'Количество банкротов: {fn}, процент банкротсва от выданных займов: {(fn / (tn + fn)) * 100:.1f}%')
```
